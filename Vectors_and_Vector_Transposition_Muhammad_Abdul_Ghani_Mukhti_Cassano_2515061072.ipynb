Tensorflow

```python
import tensorflow as tf
```

```python
x = tf.constant([[0],
                 [7],
                 [2]])
x
```

```python
x_t = tf.transpose(x)
x_t
```

```python
y = tf.constant([[6],
                 [1],
                 [1]])
```

```python
y_t = tf.transpose(y)
y_t
```
