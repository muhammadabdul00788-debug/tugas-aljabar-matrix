2. Lakukan tahapan pada soal 1 dengan menggunakan tensorflow

```python
import tensorflow as tf
```

```python
#buat variabel skalar x dan simpan suatu nilai skalar
x_tf = 72
```

```python
#tampilkan tipe data dari x
x_tf = tf.Variable(72, dtype=tf.int32)
x_tf
```

```python
#buat variabel skalar y dan simpan suatu nilai skalar
y_tf = 7
```

```python
#tampilkan tipe data dari y
y_tf = tf.Variable(7, dtype=tf.int32)
y_tf
```

```python
#jumlahkan x + y
x_tf + y_tf
```

```python
#buat variabel bernama x_float dan simpan suatu nilai desimal
tf_float = 0.7
```

```python
#tampilkan tipe data dari x_float
tf_float = tf.Variable(0.7, dtype=tf.float16)
tf_float
```

3. Lakukan tahapan pada soal 1 dengan menggunakan pytorch

```python
import torch as pt
```

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x_pt = 72
x_pt
```

```python
# tampilkan tipe data dari x
x_pt = pt.tensor(72, dtype=pt.int32)
x_pt
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y_pt = 7
y_pt
```

```python
# tampilkan tipe data dari y
y_pt = pt.tensor(7, dtype=pt.int32)
y_pt
```

```python
# jumlahkan x+y
x_pt + y_pt
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
xpt_float = 0.7
```

```python
# jumlahkan x_float dengan y
ptfloat_sum = pt.add(xpt_float, y_pt)
ptfloat_sum
```
